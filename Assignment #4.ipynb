```python
import pandas as pd
import numpy as np

#Q1. Find the total score for each student
#Q2. Find the percentage total score for each student
#Q3. Grade the students eg. Grade A, B, C & D

data = pd.read_csv('student.csv')
data
```

```python
#Q1. Find the total score for each student

data['total_score'] = (data['math score']) + (data['reading score']) + (data['writing score'])
data
```

```python
#Q2. Find the percentage total score for each student

data['percent_total_score'] = (((data['total_score'])/300)*100).round(1)
data
```

```python
#Q3. Grade the students eg. Grade A, B, C & D

def grade(x):
    if x < 50:
        return 'D'
    elif x >= 50 and x < 60:
        return 'C'
    elif x >= 60 and x < 80:
        return 'B'
    else:
        return 'A'
```

```python
grade(75)
```

```python
data['stud_grade'] = data['percent_total_score'].apply(grade)
data
```

```python
#Q4. Group students by their grades, and compute the average of each group

data.groupby(['stud_grade']).mean()
```
