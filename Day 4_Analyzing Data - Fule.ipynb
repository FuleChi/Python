```python
#Distinguish values that below 2, between 2 and 5, and above 5

def tip_rate(x):
    if x <= 2:
        return '*'
    elif x > 2 and x < 5:
        return '**'
    else:
        return '***'
```

```python
#Verify new function

tip_rate(6)
```

```python
df['tip_r'] = df['tip'].apply(tip_rate)
df
```

```python
#Merge in Pandas
reg = pd.DataFrame({'reg':[1,2,3,4], 'name':['Andrew','Chioma','Tochi','Princess']})
logins = pd.DataFrame({'login':[1,2,3,4], 'name':['Ekundayo','Andrew','Bola','Princess']})
```

```python
reg
```

```python
logins
```

```python
#Inner Join - What is common to both data

left_join_data = pd.merge(reg, logins, how='inner')
left_join_data
```

```python
#Left Join - What is common to both data + values in the left

inner_join_data = pd.merge(reg, logins, how='left')
inner_join_data
```

```python
#Right Join - What is common to both data + values in the right

right_join_data = pd.merge(reg, logins, how='right')
right_join_data
```

```python
#Outer Join - The combination of right and left data

outer_join_data = pd.merge(reg, logins, how='outer')
outer_join_data
```
